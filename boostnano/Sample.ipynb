```python
from boostnano.boostnano_model import CSM
from boostnano.boostnano_eval import evaluator
import torch
import os
import numpy as np
import h5py
from matplotlib import pyplot as plt
```

```python
#Read signal from sampled fast5 file
test_fast5 = "../sample_data/sample_0.fast5"
root = h5py.File(test_fast5)
signal = np.asarray(list(root['/Raw/Reads'].values())[0][('Signal')],dtype = np.float32)
```

```python
#Initial the evaluator
net = CSM()
ev = evaluator(net,'../model/')
(decoded,path,locs) = ev.eval_sig(signal,1000) #Segment length have to be the same when training the model, in the default model is 1000
```

```python
#Plot the result
plt.plot(signal)
for loc in locs:
    plt.axvline(x=loc,color = 'red')
```
